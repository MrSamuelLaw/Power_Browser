The important assumptions are:
- That the wheel and cylindar operate under a no slip condition. 
- That the rise time of the sensor can be neglected as it is between 2-4 micro-seconds.
- That the max speed for which power estimations are desired is 35 mph.
</pre>

```python
import math as m
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step
```

```python
# input known values
wheel_diameter = 2100e-3     # m
cylinder_diameter = 76.2e-3  # m
mps_max = 15.65              # m/s  based off of 35 mph max speed
plant_rise_time = 2          # micro_seconds

# calculate omega of wheel using v = wr 
wheel_rps = mps_max/(0.5*wheel_diameter)

# calculate omega of cylinder using |w_cylinder| = gr*|w_wheel|
cylinder_rps = wheel_rps * (wheel_diameter/cylinder_diameter)  # rad/s
cylinder_hz = cylinder_rps/(2*m.pi)                            # hz

# estimate the time the sensor will be in front
# of the shiny strip
strip_width = 6.35e-3                                          # m
flywheel_diameter = 165.1e-3                                   # m
flywheel_circumference = m.pi*flywheel_diameter                # m
t_min = (strip_width/flywheel_circumference)*(1/cylinder_hz)   # s
t_min_micros = t_min/(1e-6)                                    # micros
t_desired_micros = t_min_micros - plant_rise_time              # micros

print(f"{t_min_micros = :0.2f} micro-seconds")
print(f"{t_desired_micros = :0.2f} micro-seconds")
```

```python
R = 10     # ohms
C = 2.2e-6 # farads
G = TransferFunction([1], [R*C, 1])
t, y = step(G)
plt.plot(t*1e6, y)
plt.grid()
plt.xlabel("time [micros]")
plt.ylabel("signal strength [%]")
plt.title("Transient Response");
```

### Digital Alternative
<pre>
The alternative is to throw out readings that are clearly wrong, either by setting 
a cuttoff value, or by calculating the derivative of the signal, which is angular acceleration, 
and setting a cuttoff value for that. 
</pre>

<pre>
The current arduino code looks like such, to probe for valid times, we will implement 
a test to see if it is above the threashold, plus a small margin to account for error.
</pre>

<pre>
The new code tries twice to get a valid reading, with a threash hold value of 185.
If both attempts are unsuccessful, it will return the default value of zero.
The result of this modification was that... it worked great for lower rpm's however
after about the 30 watt mark it filtered out valid power values. 
</pre>

<pre>
With the last attempt a failure, calculating the max acceration and 
applying a cutoff filter to the signal derivative would be necessary
</pre>

```python
# calculate max detectable acclerations
v0 = 0.44704  # m/s
v1 = 15.9408   # m/s

t0 = 0.0  # s
t1 = 10.0  # s

dvdt = (v1 - v0)/(t1-t0)
print(f"{dvdt = :0.2f}  m/(s^2)")
```

```python
# solve for wheel diameter
wheel_circumference = 2100e-3                # m
wheel_radius = wheel_circumference/(2*m.pi)  # m

# solve for wheel rad/s using v = w*r
w_wheel_0 = v0/wheel_radius  # rad/s
w_wheel_1 = v1/wheel_radius  # rad/s

# solve for the cylinder rad/s
cylinder_diameter = 54e-3                                  # m
cylinder_radius = cylinder_diameter/2                      # m
w_cylinder_0 = w_wheel_0 * (wheel_radius/cylinder_radius)  # rad/s
w_cylinder_1 = w_wheel_1 * (wheel_radius/cylinder_radius)  # rad/s

# solve for cylindar time per revolution
rps_cylinder_0 = w_cylinder_0/(2*m.pi)  # rps
rps_cylinder_1 = w_cylinder_1/(2*m.pi)  # rps

seconds_0 = 1/rps_cylinder_0  # s
seconds_1 = 1/rps_cylinder_1  # s

micros_0 = seconds_0/(1e-6)
micros_1 = seconds_1/(1e-6)

dsdt = (micros_1 - micros_0)/(t1 - t0)
print(f"{dsdt = :0.2f}  micros/sec")
print(micros_1)
```

```python
# input coeffecients
a = 199_844_880_469.07413 - 48_000_000_000;
b = -7_111_362.674411774 + 2_100_500;
c = 68.65055109403046;

# calculate the powers
omega0 = 1/micros_0
omega1 = 1/micros_1
p0 = (a*(omega0**2)) + (b*omega0) + c
p1 = (a*(omega1**2)) + (b*omega1) + c
print(f"{p0 = :0.2f} W")
print(f"{p1 = :0.2f} W")
```
